```python
merged_dict = {}
for school, table in dfs.items():
    merged_dict[school] = pd.merge(table, mon, on=['school', 'subject', 'name'])
    merged_dict[school] = merged_dict[school].dropna()
    print(table.isna().value_counts())
```

```python
for school, table in merged_dict.items():
    print(table.isna().values.any())
```

```python
from sklearn.preprocessing import MinMaxScaler
# Since the dependent and independent variable have different max scales, we need to Normalize the data first
for school, table in merged_dict.items():
        table.loc[:, ['soch2_norm']] = MinMaxScaler().fit_transform(table.loc[:, ['soch2']]).round(2)
        table.loc[:, 'soch4_norm'] = MinMaxScaler().fit_transform(table.loc[:, ['soch4']]).round(2)
        table.loc[:, 'RS1122_norm'] = MinMaxScaler().fit_transform(table.loc[:, ['RS1122']]).round(2)
        table.loc[:, 'RS0423_norm'] = MinMaxScaler().fit_transform(table.loc[:, ['RS0423']]).round(2)
```

```python
dfs_grade = {}
grade_list = [4, 8]
for school, table in merged_dict.items():
    grade = {}
    
    for g in grade_list:
        filtered_table = table[table['subject'].str.contains(str(g))]
        if not filtered_table.empty:
            grade[g] = filtered_table
    
    dfs_grade[school] = grade
```

# Regression analysis

---

```python
import statsmodels.api as sm
```

```python
dfs_grade['КГУ «Бестамакская средняя школа» Алгинского района Актюбинской области'][8]
```

* Term 2

```python
for school, grades in dfs_grade.items():
    for grade, t2 in grades.items():
        X = t2['RS1122_norm']
        y = t2['soch2_norm']

        X = sm.add_constant(X)
        model = sm.OLS(y, X).fit()
        print(f'{school} {grade} Term 2\n')
        print(f'{model.summary()}\n\n')
```

* Term 4

```python
for school, grades in dfs_grade.items():
    for grade, t4 in grades.items():
        X = t4['RS0423_norm']
        y = t4['soch4_norm']

        X = sm.add_constant(X)
        model = sm.OLS(y, X).fit()
        print(f'{school} {grade} Term 4\n')
        print(f'{model.summary()}\n\n')
```

* Term 2 + Term 4

```python
for school, grades in dfs_grade.items():
    for grade, tAll in grades.items():
    # Combining the 2 terms
        x1 = tAll['RS1122_norm']
        x2 = tAll['RS0423_norm']
        y1 = tAll['soch2_norm']
        y2 = tAll['soch4_norm']
        
        X = pd.DataFrame(pd.concat([x1, x2]), columns=['Мониторинг'])
        y = pd.DataFrame(pd.concat([y1, y2]), columns=['СОЧ'])

        X = sm.add_constant(X)
        model = sm.OLS(y, X).fit()
        print(f"{school} All Terms\n")
        print(f"{model.summary()}\n\n")
```

* By subjects

```python
for school, subjects in school_grade_subject.items():
    for subject, table in subjects.items():
    # Combining the 2 terms
        x1 = table['RS1122_norm']
        x2 = table['RS0423_norm']
        y1 = table['soch2_norm']
        y2 = table['soch4_norm']
        
        X = pd.DataFrame(pd.concat([x1, x2]), columns=['Мониторинг'])
        y = pd.DataFrame(pd.concat([y1, y2]), columns=['СОЧ'])

        X = sm.add_constant(X)
        model = sm.OLS(y, X).fit()
        print(f"{school} {subject}\n")
        print(f"{model.summary()}\n\n")
```
