```python
schools = [
    'КГУ «Общеобразовательная школа имени Досова» с. Катарколь Бурабайского района Акмолинской области', 
    'КГУ «Бестамакская средняя школа» Алгинского района Актюбинской области',
    'КГУ «Общеобразовательная школа № 5 города Ақсай» Бурлинского района ЗКО'
]
```

```python
# def sliced + percent multiplied by 100
df = df.loc[df['school_ru'].isin(schools), ['test', 'school_ru', 'fio', 'subject', 'percent']].rename(columns={'school_ru' : 'school', 'fio' : 'name'})
df['percent'] = df['percent'] * 100
```

```python
mon = df.pivot_table(index=['school', 'name', 'subject'], columns='test', values='percent').reset_index()
mon
```

# Joining tables

---

```python
# aqmola = pd.DataFrame()
# aqtobe = pd.DataFrame()
# west = pd.DataFrame()

merged_dict = {}
for school, table in dfs.items():
    merged_dict[school] = pd.merge(table, mon, on=['school', 'subject', 'name'])
```

# Regression analysis

---

```python
import statsmodels.api as sm
```

* Term 2

```python
for school, table in merged_dict.items():
    X = table[~(table['soch2'].isna() | table['RS1122'].isna())]['RS1122']
    y = table[~(table['soch2'].isna() | table['RS1122'].isna())]['soch4']

    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    print(f'{school}\n')
    print(f'{model.summary()}\n\n')
```

* Term 4

```python
for school, table in merged_dict.items():
    X = table[~(table['soch4'].isna() | table['RS0423'].isna())]['RS0423']
    y = table[~(table['soch4'].isna() | table['RS0423'].isna())]['soch4']

    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    print(f'{school}\n')
    print(f'{model.summary()}\n\n')
```
